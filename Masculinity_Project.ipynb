## Mapping the Data
***

In order for us to start thinking about using the KMeans algorithm with this data, we need to first figure out how to turn these responses into numerical data. Let's once again consider question 7. We can't cluster the data using the phrases `"Often"` or `"Rarely"`, but we can turn those phrases into numbers. We could map the data in the following way: 
* `"Often"` -> `4`
* `"Sometimes"` ->  `3`
* `"Rarely"` -> `2` 
* `"Never, but open to it"` -> `1`
* `"Never, and not open to it"` -> `0`.



```python
cols_to_map = ["q0007_0001", "q0007_0002", "q0007_0003", "q0007_0004",
       "q0007_0005", "q0007_0006", "q0007_0007", "q0007_0008", "q0007_0009",
       "q0007_0010", "q0007_0011"]
```

```python
for col in cols_to_map:
    survey[col]=survey[col].map({"Often":4,"Rarely":2,"Sometimes":3,"Never, but open to it":1,"Never, and not open to it":0})

survey['q0007_0001']
```

```python
survey['q0007_0001'].value_counts()
# We can see that the mapping preserve the values count, then it's done successfully
```

```python
## Mapping question 29
col = survey['q0029']
```

```python
survey['q0029'].value_counts()
```

```python
col = col.map({"Post graduate degree":5,"College graduate":4,"Some college":3,"Associate's degree":2,"High school or G.E.D.":1,"Did not complete high school":0})
col.value_counts()
```

# Build the KMeans Model
***
KMeans is already availablein sklearn, We'll import it from their 

```python
from sklearn.cluster import KMeans
```

```python
rows_to_cluster = survey.dropna(subset=["q0007_0001", "q0007_0002", "q0007_0003", "q0007_0004", "q0007_0005", "q0007_0008", "q0007_0009"])
rows_to_cluster
```

```python
model = KMeans(n_clusters=2, init='random')
model.fit(rows_to_cluster[["q0007_0001", "q0007_0002", "q0007_0003", "q0007_0004", "q0007_0005", "q0007_0008", "q0007_0009"]])
print(model.labels_)
print(model.cluster_centers_)
```

```python
cluster_zero_indices = []
cluster_one_indices = []
cluster_zero_indices, cluster_one_indices = [i for i, label in enumerate(model.labels_) if label == 0], [i for i, label in enumerate(model.labels_) if label == 1]
print(cluster_one_indices)
print(cluster_zero_indices)
```

```python
# People in cluster 0
cluster_zero_df = rows_to_cluster.iloc[cluster_zero_indices]
cluster_zero_df
```

```python
# People on cluster 1
cluster_one_df = rows_to_cluster.iloc[cluster_one_indices]
cluster_one_df
```

```python
# Explore the clusters 
print(cluster_one_df['educ4'].value_counts())
print(cluster_zero_df['educ4'].value_counts())
print(cluster_one_df['age3'].value_counts())
print(cluster_zero_df['age3'].value_counts())
```
